```python
# Inicializar una lista vicia.
datos = []

# Funciones del sistema

# funcion agregarDatos
def agregarDatos():
    while True:
        dato = input("Ingrese un nuevo datos para la lista, teclee '0' para salir: ")
        if dato == "0":
            break
        datos.append(dato)

def publicarLista():
    if len(datos) == 0:
        print("La lista esta vacia")
    for dato in datos:
        print(dato)

def eliminarDatos():
    if len(datos) != 0:
        while True:
            dato = input("Ingrese el dato que va a eliminar, teclee '0' para salir: ")
            if dato == "0":
                break
            elif dato == "":
                print("Teclee un dato a eliminar, por favor.")
                continue
            else:
                if datos.count(dato) == 0:
                    print(f"Este dato: '{dato}' no se encuentra, intente de nuevo por favor.")
                else:
                    datos.remove(dato)
                    print(f"El dato: '{dato}', fue eliminado con exito.")
    else:
        print("La aplicaci贸n se cerr贸.")

while True:
    menu = """
            MENU PRINCIPAL
        1) Agregar datos a la lista.
        2) Modificar datos de la lista.
        3) Eliminar datos de la lista.
        4) Imprimir la lista
        5) Salir.
        Escoja una opci贸n: 
    """
          
    opcion = int(input(menu))
    if opcion == 1:
        agregarDatos()
    elif opcion == 2:
        pass
    elif opcion == 3:
        eliminarDatos()
    elif opcion == 4:
        publicarLista()  
    elif opcion == 5:
        break
    else:
        print("Escojio una opci贸n incorrenta, intente de nuevo por favor.")
```
